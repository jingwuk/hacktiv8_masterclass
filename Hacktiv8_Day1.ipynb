Topik: Eksplorasi Lagu Populer di Indonesia
Pertanyaan 1: Apa saja lagu populer di Indonesia

```python
#Step 1 : Ambil semua data lagu dari country == Indonesia

data_lagu_indonesia = dataset[dataset['country']=='indonesia']
data_lagu_indonesia

#Step 2 : Cari genre lagu

data_lagu_indonesia['genre'].value_counts()

#Step 3 : Visualisasi (karena komposisi maka pakai piechart)
#memasukkan jumlah genre ke dataset dan reset index
data_genre = data_lagu_indonesia['genre'].value_counts().to_frame().reset_index()
#rename kolom 
data_genre.columns = ['genre','jumlah']
data_genre
```

```python
#Step 4 : Visualisasi
data_genre.plot(kind='pie',
                labels=data_genre['genre'],
                y='jumlah',
                legend=False,
                figsize=(5,5))
```

```python
#Step 5 : Cleaning Visualization

data_genre_top_5 = data_genre[0:5]
data_others = {
    'genre':'others',
    'jumlah':data_genre[5:]['jumlah'].sum()
}
data_genre_top_5 = data_genre_top_5.append(data_others, ignore_index=True)
data_genre_top_5
```

```python
#Step 6 : Visualisasi Final
data_genre_top_5.plot(kind='pie',
                labels=data_genre_top_5['genre'],
                y='jumlah',
                legend=False,
                figsize=(5,5),
                #berikan persentase
                autopct='%1.1f%%',
                title='Top 5 Genre Populer di Indonesia')
```

Pertanyaan 2 : Orang-orang indonesia  cenderung galau? Benarkah demikian?

```python
#Step 1 : dapatkan data Valence
data_valence = dataset.groupby('country')['valence'].mean().sort_values(ascending=False)
data_valence
```

```python
#Step 2 : Ubah ke dataframe

data_valence = data_valence.to_frame().reset_index()
data_valence.columns = ['country','mean_valence']
data_valence
```

```python
#step 3 : visualisasi

data_valence.plot(kind='barh',
                  x='country',
                  y='mean_valence',
                  figsize=(7,7),
                  title='Perbandingan Mean Valence Antar Negara')
```

Narasi : ingin mencari tren popularity dari Artist Jepang bernama Aimyon

```python
#Step 1 : Dapatkan dataset
dataset_aimyon = dataset[ (dataset['country']=='japan')&
                          (dataset['artist']=='Aimyon')].groupby('year')['popularity'].mean()
dataset_aimyon
```

```python
#Step 2 : Ubah ke dataframe
dataset_aimyon = dataset_aimyon.to_frame().reset_index()
dataset_aimyon.columns = ['Tahun','Mean_Popularity']
dataset_aimyon
```

```python
#Step 3 : Visualisasi
dataset_aimyon.plot(kind='line',
                    x='Tahun',
                    y='Mean_Popularity',
                    figsize=(10,5))
```

```python
import matplotlib.pyplot as plt
```

```python
#Step 4 : Clean Visualization
dataset_aimyon['Tahun'] = dataset_aimyon['Tahun'].astype('category')
dataset_aimyon.plot(kind='line',
                    x='Tahun',
                    y='Mean_Popularity',
                    figsize=(10,5))
plt.ylim(ymin=0, ymax=100)
plt.title('Tren Aimyon')
```
